```python
import os
import csv

path = './task/train'
dir_list = os.listdir(path)

# CSV 파일 생성 및 헤더 추가
with open('task_train_info.csv', 'w', newline='') as csvfile:
    fieldnames = ['path', 'gender', 'age']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for file_name in dir_list:
        gender = 1 if 'F' in file_name else 0
        parts = file_name.split('_')
        age = parts[-2]

        # 파일 이름과 성별 정보를 CSV 파일에 쓰기
        writer.writerow({'path': './dataset/task/train/'+file_name, 'gender': gender, 'age': age})

print("CSV 파일이 생성되었습니다.")
```

```python
import os
import csv

path = './task/valid'
dir_list = os.listdir(path)

# CSV 파일 생성 및 헤더 추가
with open('task_valid_info.csv', 'w', newline='') as csvfile:
    fieldnames = ['path', 'gender', 'age']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for file_name in dir_list:
        gender = 1 if 'F' in file_name else 0
        parts = file_name.split('_')
        age = parts[-2]

        # 파일 이름과 성별 정보를 CSV 파일에 쓰기
        writer.writerow({'path': './dataset/task/valid/'+file_name, 'gender': gender, 'age': age})

print("CSV 파일이 생성되었습니다.")
```

```python
import pandas as pd

df = pd.read_csv('task_train_info.csv')
print(df.head)

row_count = len(df)
print("행 개수:", row_count)

gender_count = df['gender'].value_counts()
print(gender_count)
print("0이 남자, 1이 여자입니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('task_train_info.csv')
age_gender_count = df.groupby(['age', 'gender']).size().unstack(fill_value=0)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(14, 6))
age_gender_count.plot(kind='bar', stacked=True, color=['blue', 'red'], ax=ax)
ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.set_title('Count of Data by Age and Gender')
ax.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```
